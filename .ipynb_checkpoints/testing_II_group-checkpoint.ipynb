```python
%reload_ext autoreload

import pandas as pd

import numpy as np

from sklearn.linear_model import LogisticRegression

from make_ready_data.make_ready_big_data import read_data, get_label, get_feat

from make_ready_data.make_ready_big_data import look_through_info, get_train_and_test_sets

from make_ready_data.make_ready_short_data import get_short_test_data_from_Roman

from make_ready_data.make_ready_short_data import get_short_test_data_from_Svetlana_G

from make_ready_data.make_ready_short_data import extraxt_samples_from_new_test_data

from make_ready_data.create_features_indeces import get_labels_and_features_indeces

from make_ready_data.create_features_indeces import build_plot_for_best_features
```

```python
info, final_filelist = read_data()

data_sick, data_heal, inds_sick, inds_heal = look_through_info(info, final_filelist)

plot_labels, indlist = get_labels_and_features_indeces()
```

```python
test_patients_data = get_short_test_data_from_Svetlana_G()
test_patients_data.append(('Roman_patient', get_short_test_data_from_Roman()))
```

```python
train_sick_samples = extraxt_samples_from_new_test_data(test_patients_data)
```

```python
labels = np.ones((train_sick_samples.shape[0], 1))
train_sick_samples_with_labels = np.concatenate((train_sick_samples, labels), axis=1)
```

```python
def testing_all_new_test_samples(data_sick, data_heal, indlist, test_patients_samples):
    all_data = np.concatenate((data_heal, data_sick))
    res = np.zeros(len(indlist))
    for i, inds in enumerate(indlist):
        clf = LogisticRegression(C=1000.)
        clf.fit(all_data[:, inds], all_data[:, -1])
        res[i] = clf.predict(test_patients_samples[:, inds]).sum() / test_patients_samples.shape[0]
    return res
```

```python
def form_new_test_patients(all_patients_data):
    new_test_patients_data = []
    for i, data in enumerate(all_patients_data):
        if data[0, -1] == 0:
            continue
        new_test_patients_data.append(('patient'+str(i), data))
    return new_test_patients_data
```

```python
data_sick, data_heal, inds_sick, inds_heal = look_through_info(info, final_filelist)
_, _, _, _, old_patients_data = get_train_and_test_sets(data_sick, data_heal, inds_sick, inds_heal, 
                                                   per_edge=0, balanced_data=False, make_shuffle=False)
new_test_patients_data = form_new_test_patients(old_patients_data)
```

```python
def testing_all_new_test_patients(data_sick, data_heal, indlist, test_patients_data):
    all_data = np.concatenate((data_heal, data_sick))
    res_for_patients = np.zeros(len(indlist))
    for i, inds in enumerate(indlist):
#         print(inds)
        clf = LogisticRegression(C=1000.)
        clf.fit(all_data[:, inds], all_data[:, -1])
        for j in range(len(test_patients_data)):
            res = clf.predict(test_patients_data[j][1][:, inds])
            log_ans = round(res.sum() / res.shape[0] + 0.0001)
#             print(test_patients_data[j][0], int(log_ans), res)
            res_for_patients[i] += log_ans
#         print()
    
    res_for_patients = res_for_patients / len(test_patients_data)
    return res_for_patients
```

```python
res_for_samples = testing_all_new_test_samples(train_sick_samples_with_labels, 
                                               data_heal[:90], indlist, data_sick[:, :-1])
res_for_patients = testing_all_new_test_patients(train_sick_samples_with_labels,
                                                 data_heal[:90], indlist, new_test_patients_data)
```

```python
n_heals = 92
n_tests = 500
aver_res_for_samples = np.zeros(len(indlist))
aver_res_for_patients = np.zeros(len(indlist))
for _ in range(n_tests):
    if n_tests%100 == 0:
        print(n_tests)
    i1 = np.random.randint(0, data_heal.shape[0]-n_heals)
    i2 = i1 + n_heals
    aver_res_for_samples += testing_all_new_test_samples(train_sick_samples_with_labels, 
                                               data_heal[i1:i2], indlist, data_sick[:, :-1])
    aver_res_for_patients += testing_all_new_test_patients(train_sick_samples_with_labels,
                                                 data_heal[i1:i2], indlist, new_test_patients_data)
aver_res_for_samples = aver_res_for_samples / n_tests
aver_res_for_patients = aver_res_for_patients / n_tests
```

```python
build_plot_for_best_features(plot_labels, indlist,
                             y=[aver_res_for_samples, aver_res_for_patients],
                             y_arange=np.arange(0.0, 1.10 ,0.1),
                             marker=[':ro', ':go'],
                             legend_label=['  , C=1000', 
                                           '  , C=1000'], 
                             title=str(n_heals)+'  c  '+',  ', 
                             savefile_name=str(n_heals)+' samples'+'_average')
```
