```python
def cross_validation_on_all_data(DataSick, DataHeal, plot_labels, indlist, C=1000.):
        
    data_sick = DataSick.copy()
    data_heal = DataHeal.copy()
            
    all_data = np.concatenate((data_heal, data_sick))
    np.random.seed(0)
    np.random.shuffle(all_data)
    
    res_for_all_features = np.zeros(len(indlist))

    for i in range(len(indlist)):
        new_all_data = all_data[:, indlist[i]+[-1]]
        print(indlist[i])
        for j in range(all_data.shape[0]):
            train_data = np.delete(new_all_data, j, 0)
            test_data = new_all_data[j]
            clf = LogisticRegression(C=C)
            clf.fit(train_data[:, :-1], train_data[:, -1])
            log_res = clf.predict(test_data[:-1].reshape(1, -1))[0]
            res_for_all_features[i] += test_data[-1] == log_res
    res_for_all_features = res_for_all_features / all_data.shape[0]
    return res_for_all_features
```

```python
y_100 = cross_validation_on_all_data(data_sick, data_heal, plot_labels, indlist, C=100)
y_1000 = cross_validation_on_all_data(data_sick, data_heal, plot_labels, indlist, C=1000)
```
