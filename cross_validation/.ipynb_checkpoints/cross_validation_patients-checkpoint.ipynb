```python
%reload_ext autoreload
import pandas as pd

import numpy as np

from sklearn.linear_model import LogisticRegression

from make_ready_data.make_ready_big_data import read_data, get_label, get_feat

from make_ready_data.make_ready_big_data import look_through_info, get_train_and_test_sets

from make_ready_data.create_features_indeces import get_labels_and_features_indeces

from make_ready_data.create_features_indeces import build_plot_for_best_features
```

```python
info, final_filelist = read_data()
data_sick, data_heal, inds_sick, inds_heal = look_through_info(info, final_filelist)
```

```python
plot_labels, indlist = get_labels_and_features_indeces()
```

```python
_, _, _, _, patient_data = get_train_and_test_sets(data_sick, data_heal, inds_sick, inds_heal, 
                                                   per_edge=0, balanced_data=False)
```

```python
def cross_validation_on_all_patients(patient_data, indlist, C=1000.):
    res_for_all_features = np.zeros(len(indlist))
    def form_train_data(patient_data):
        train_data = np.empty((0, 6))
        for data in patient_data:
            train_data = np.concatenate((train_data, data))
        return train_data
        
    for i, inds in enumerate(indlist):
        print(inds)
        for j in range(len(patient_data)):
            patient_data_copy = patient_data.copy()
            test_data = patient_data_copy.pop(j)[:, inds+[-1]]
            train_data = form_train_data(patient_data_copy)[:, inds+[-1]]            
            clf = LogisticRegression(C=C)
            clf.fit(train_data[:, :-1], train_data[:, -1])
            log_res = round(clf.predict(test_data[:, :-1]).sum() / test_data.shape[0])
            res_for_all_features[i] += test_data[0, -1] == log_res
    res_for_all_features = res_for_all_features / len(patient_data)
    return res_for_all_features
```

```python
y_1000 = cross_validation_on_all_patients(patient_data, indlist, C=1000)
```

```python
build_plot_for_best_features(plot_labels, indlist, 
                             y=[y_1000],
                             y_arange=np.arange(0.55,0.75,0.01),
                             marker=[':ro', ':go'],
                             legend_label=['-  , C=1000', 
                                           '   1000 , C=1000'])
```
