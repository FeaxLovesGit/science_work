```python
%reload_ext autoreload
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2, f_classif, mutual_info_classif, f_regression
from sklearn.decomposition import PCA
from sklearn.utils import shuffle
from sklearn import svm as SVM
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.svm import SVC
import scipy.io
import os
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from pylab import plot, show
import matplotlib.pyplot as plt
import matplotlib.markers as mmark
import matplotlib.lines as mlines
from matplotlib.pyplot import  scatter
from sklearn.model_selection import train_test_split
from make_ready_big_data import read_data, get_label, get_feat
from make_ready_big_data import look_through_info, get_train_and_test_sets
from mpl_toolkits import mplot3d
```

```python
info, final_filelist = read_data()
data_sick, data_heal, inds_sick, inds_heal = look_through_info(info, final_filelist)
```

```python
# plot_labels = ['RR', 'TQ', 'QTc', 'JTc', 'TpeakTend']
plot_labels = ['QTc', 'JTc', 'TpeakTend', 'TQ', 'RR']
N_FEATURES = len(plot_labels)
```

```python
indlist = []
for i in range(1, 2**N_FEATURES):
    indlist_tmp = []
    bin_str = '{0:b}'.format(i)
    if len(bin_str) < 5:
        bin_str = '0'*(5-len(bin_str)) + bin_str

    for j,c in enumerate(bin_str):
        if c=='1':
            indlist_tmp.append(N_FEATURES-j-1)
    indlist.append(indlist_tmp)
```

```python
n_tests = 1000
log_res_10 = np.zeros(len(indlist))
log_res_100 = np.zeros(len(indlist))
log_res_1000 = np.zeros(len(indlist))

for i in range(len(indlist)):
    print('indlist:', indlist[i])
    for _ in range(n_tests):
        tr_d, ts_d, tr_l, ts_l = get_train_and_test_sets(data_sick, data_heal, inds_sick, inds_heal)
        
        new_tr_d = tr_d[:, indlist[i]]
        new_ts_d = ts_d[:, indlist[i]]

        clf = LogisticRegression(C=10.)
        clf.fit(new_tr_d, tr_l)
        res = clf.score(new_ts_d, ts_l)
        log_res_10[i] += res

        clf = LogisticRegression(C=100.)
        clf.fit(new_tr_d, tr_l)
        res = clf.score(new_ts_d, ts_l)
        log_res_100[i] += res
        
        clf = LogisticRegression(C=1000.)
        clf.fit(new_tr_d, tr_l)
        res = clf.score(new_ts_d, ts_l)
        log_res_1000[i] += res

    
log_res_10 = log_res_10 / n_tests
log_res_100 = log_res_100 / n_tests
log_res_1000 = log_res_1000 / n_tests
```

```python
# np.savetxt("LogisticRegression_1000", log_res_1000)
```

```python
short_indlist = []
short_logres1000 = []
short_labels = []
for i in range(len(indlist)):
    if len(indlist[i]) <= 2:
        short_indlist.append(indlist[i])
        short_logres1000.append(log_res_1000[i])
        short_labels.append(str(np.array(plot_labels)[indlist[i]]))
```

```python
# построене графика для всех комбинаций с не более чем двумя признаками

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15,10)
n = len(short_indlist)
x = np.arange(n)
plt.plot(x, short_logres1000, ':bo', label='LogisticRegression: C = 1000')

# plt.legend(fontsize=15)

plt.xticks(range(n), [short_labels[i] for i in range(n)], fontsize=12, rotation=90)
plt.yticks(np.arange(0.53,0.7,0.005), fontsize=15)

plt.xlabel('Features', fontsize=15)
plt.ylabel('Score', fontsize=15)
plt.savefig('short_find_best_features.png', bbox_inches='tight')
plt.show()
```
